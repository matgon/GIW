```python
import urllib
import re
from bs4 import BeautifulSoup
from pathlib import Path

html=urllib.request.urlopen("https://es.wikipedia.org/wiki/Anexo:Cuadros_de_Pedro_Pablo_Rubens").read()
soup=BeautifulSoup(html,'html.parser')
```

```python
tablas = soup("table")
num_pinturas = 0
lista_pinturas = list()

CLEANR = re.compile('<.*?>')
#función para limpiar el html
def _cleanhtml(raw_html):
    cleantext = re.sub(CLEANR, '', raw_html)
    return cleantext

#para rellenar la lista de pinturas, donde cada pintura es un diccionario
for i in range(1, 7):
    filas_tabla = tablas[i]("tr")

    for j in range(1, len(filas_tabla)):
        try:
            fila_contents = filas_tabla[j].contents
            lista_pinturas.append(dict())
            try:
                lista_pinturas[num_pinturas]["image"] = fila_contents[1].contents[0].contents[0].get("src", None)
            except:
                lista_pinturas[num_pinturas]["image"] = None
            try:
                lista_pinturas[num_pinturas]["title"] = _cleanhtml(str(fila_contents[3]))
            except:
                lista_pinturas[num_pinturas]["title"] = None               
            try: 
                lista_pinturas[num_pinturas]["dimensions"] = _cleanhtml(str(fila_contents[5]))
            except:
                lista_pinturas[num_pinturas]["dimensions"] = None
            try:
                lista_pinturas[num_pinturas]["date"] = _cleanhtml(str(fila_contents[7]))
            except:
                lista_pinturas[num_pinturas]["date"] = None
            try:
                lista_pinturas[num_pinturas]["material"] = _cleanhtml(str(fila_contents[9]))
                if lista_pinturas[num_pinturas]["material"] == "\n":
                    lista_pinturas[num_pinturas]["material"] = None
            except:
                lista_pinturas[num_pinturas]["material"] = None
            try:
                lista_pinturas[num_pinturas]["references"] = fila_contents[11].contents[0].get("href", None)
            except:
                lista_pinturas[num_pinturas]["references"] = None
            try:
                try:
                    lista_pinturas[num_pinturas]["localization"] = _cleanhtml(str(fila_contents[13]))
                except:
                    lista_pinturas[num_pinturas]["localization"] = None
            except:
                lista_pinturas[num_pinturas]["localization"] = None
            num_pinturas += 1
        except:
            num_pinturas += 1
```

```python
# Usaremos una lista para cada etapa
listEtapa1 = list()
listEtapa2 = list()
listEtapa3 = list()
listEtapa4 = list()
listEtapa5 = list()

# Funcion para guardar las pinturas que han sido elegidas por el usuario en una carpeta
def guardar_pinturas_etapa(name_folder: str, list_pinturas):
    folder = Path(".") / name_folder
    folder.mkdir(parents=True, exist_ok=True)

    for i in range(0, len(list_pinturas)):
      try:
        img = urllib.request.urlopen(str("https:"+list_pinturas[i]["image"])).read()
        manf=open(Path(folder) / str(list_pinturas[i]["title"] + ".jpg"),"wb")
        manf.write(img)
        manf.close()
      except:
        i -= 1

# Función que dada una pintura imprime sus campos
def print_pinturas(pinturas):
    for i in pinturas:
        print("-----------------------------------------------------------------------------------")
        try:
          print("Título: " + i['title'])
        except:
          print("Título: Desconocido")
        try:
          print("Dimensiones: " + i['dimensions'])
        except:
          print("Dimensiones: Desconocido")
        try:
          print("Fecha: " + i['date'])
        except:
          print("Fecha: Desconocido")
        try:
          print("Localización/Colección: " + i['localization'])
        except:
          print("Localización/Colección: Desconocido")
        try:
          print("Referencias: " + i['references'])
        except:
          print("Referencias: Desconocido")
        print("-----------------------------------------------------------------------------------")

# Funcion que dependiendo de la  etapa seleccionada por el usuario hace un recorrido de las pinturas y las guarda en la lista y una carpeta,
# posteriormente las imprime haciendo uso de las funciones de antes
def guardaryMostrar(sel,indice1,indice2,listaPint):
  for i in range(indice1):
    listaPint.append(lista_pinturas[i+indice2])
  guardar_pinturas_etapa(str(soup.find_all("li")[sel].find_all("span")[1].contents[0]), listaPint)
  print_pinturas(listaPint)
  

def printPinturaEtapa(selec):
  if selec == 1:
    guardaryMostrar(selec,28,5,listEtapa1)
  elif selec == 2:
    guardaryMostrar(selec,88,33,listEtapa2)
  elif selec == 3:
    guardaryMostrar(selec,94,121,listEtapa3)
  elif selec == 4:
    guardaryMostrar(selec,128,215,listEtapa4)
  else:
    guardaryMostrar(selec,71,343,listEtapa5)   

# Función principal para la primera opción, mostrar por etapa pictórica
def func1():  
    etapas = soup.find_all("li")
    for i in range(5):
      etap = etapas[i+1].find_all("span")[1].contents[0]
      print(str(i+1) + "- " + etap)
      
    print("Selecciona una etapa")
    etselec = input()
    while(int(etselec) < 1 or int(etselec) > 5):
        etselec = input()
    printPinturaEtapa(int(etselec))
```

```python
def _get_materiales_pinturas(dict_pinturas: list)-> list:
    list_materials = list()
    pinturas_material = dict()
    pinturas_material["none"] = []

    for pintura in dict_pinturas:
        try:
            aux = str(pintura["material"])
            mat = str(aux.replace("\n", "")).lower()
            if mat == None:
                pinturas_material["none"].append(pintura) # Group by unknow material
            if mat != None and mat not in list_materials: # Check if material exist
                list_materials.append(mat) # Add material to the list
                pinturas_material[mat] = [pintura]
            elif mat != None:
                pinturas_material[mat].append(pintura)
        except:
            pinturas_material["none"].append(pintura) # Group by unknow material

    return [list_materials, pinturas_material]

def _print_materiales(list_materials: list):
    i = 1
    for mat in list_materials:
        print(str(i)+"-"+mat)
        i += 1

# Título,Dimensiones, Fecha de creación, Referencia, Colección/localización
def _print_pintura_op2(pintura):
    try:
        titulo = pintura["title"]
    except:
        titulo = "desconocido"
    try:
        dimensiones = pintura["dimensions"]
    except:
        dimensiones = "desconocidas"
    try:
        creacion = pintura["creation"]
    except:
        creacion = "desconocida"
    try:
        referencia = pintura["reference"]
    except:
        referencia = "none"
    try:
        localizacion = pintura["location"]
    except:
        localizacion = "desconocida"

    print("-----------------------------------------------------------------------------------")
    print("Título: " + titulo)
    print("Dimensiones: " + dimensiones)
    print("Fecha: " + creacion)
    print("Localización/Colección: " + localizacion)
    print("Referencias: " + referencia)
    print("-----------------------------------------------------------------------------------")    

def _print_pinturas_materials(pinturas_material: list):    
    for pintura in pinturas_material:
        _print_pintura_op2(pintura)

def _save_pictures_material(name_folder: str, list_pinturas: list):
    folder = Path(".") / "materiales"
    sub_folder = Path(folder) / name_folder
    folder.mkdir(parents=True, exist_ok=True)
    sub_folder.mkdir(parents=True, exist_ok=True)

    for pintura in list_pinturas:
        if pintura['image'] != None:
            img = urllib.request.urlopen(str("https:" + pintura["image"])).read()
            manf=open(Path(sub_folder) / str(pintura['title'].strip("\n(?)") + ".jpg"),"wb")
            manf.write(img)
            manf.close()
    

def func2(dict_pinturas: list):
    [list_materials, pinturas_material] = _get_materiales_pinturas(dict_pinturas)
    _print_materiales(list_materials)
    m = int(input("Seleccione el tipo del material (0 -> Salir): "))
    if m > 0:
      print("\n", list_materials[m-1])
      _print_pinturas_materials(pinturas_material[list_materials[m-1]])
      _save_pictures_material(list_materials[m-1], pinturas_material[list_materials[m-1]])
```

```python
lista_museos = list()

#para crear una lista de museos (localizaciones)
for pintura in lista_pinturas:
    localization = pintura.setdefault("localization", None)
    if  pintura["localization"] != None:
        if pintura["localization"] not in lista_museos:
            lista_museos.append(pintura["localization"])
```

```python
import os
import urllib.request

#función que imprime la lista de museos (localizaciones)
def print_museos():
    cont = 1
    for i in lista_museos:
        print(str(cont) + ".- " + i)
        cont += 1

#función que imprime las pinturas para un museo (localización) dado
def pinturas_por_museo(museo):
    lista_pinturas_por_museo = list()
    for i in lista_pinturas:
        if i['localization'] == museo:
            lista_pinturas_por_museo.append(i)
    
    return lista_pinturas_por_museo

#función que imprime las pinturas y sus atributos            
def print_pinturas(pinturas):
    #Título,Dimensiones, Fecha de creación, Referencia, Colección/localización.
    for i in pinturas:
        print("-----------------------------------------------------------------------------------")
        print("Título: " + i['title'])
        print("Dimensiones: " + i['dimensions'])
        print("Fecha: " + i['date'])
        print("Localización/Colección: " + i['localization'])
        print("Referencias: " + i['references'])
        print("-----------------------------------------------------------------------------------")

#función para guardar las imagenes de las pinturas de un museo en una carpeta
def guardar_pinturas(lista_pinturas_por_museo):
    carpeta_inicial = os.getcwd()
    carpeta = str(str(lista_pinturas_por_museo[0]['localization']).strip("\n"))
    if not os.path.exists(carpeta):
        os.mkdir(carpeta)
        os.chdir(carpeta)
        for i in lista_pinturas_por_museo:
            if i['image'] != None:
                img = urllib.request.urlopen("https:" + i["image"]).read()
                manf=open(i['title'].strip("\n(?)") + ".jpg","wb")
                manf.write(img)
                manf.close()
        os.chdir(carpeta_inicial)

#función que realiza la lógica de la opción 3
def func3():
    print_museos()
    opt = int(input("Seleccione un museo (0 para salir): "))
    if opt > 0:
        museo = lista_museos[opt - 1]
        lista_pinturas_por_museo = pinturas_por_museo(museo)
        print_pinturas(lista_pinturas_por_museo)
        guardar_pinturas(lista_pinturas_por_museo)
```
